# Character Creator Code

```python
# SAVEPOINTCreate a function to make your character
def characterBuilder():

  # Create a global dictionary with keys labelled for character creation aspects
  # global marker and use via: https://www.w3schools.com/python/python_variables_global.asp
  global characterSubDict
  characterSubDict = {"Type" : "", "Name" : "", "Species" : "", "Traits" : "", "Flaws": "", "Backstory" : ""}
  
  # create global variable for later navigation
  global saveToDict
  saveToDict = ""

  # Welcoem them to the character maker
  print("Welcome to my character creator! Thanks for using it. \n Things should be pretty easy to use - I'll help you make a character built from your input and my suggestions. \n When I ask for your input, please hit enter after you type your answer. \n Sometimes I'll need a moment to think before we move on, so please be patient with me. \n\n That's all, let's get started! :)")

# While/return/continue/break structure via https://stackoverflow.com/questions/27364956/looping-back-to-specific-point-in-code
  # Character maker full loop
  while True:
    # Character type loop
    while True: 
    # Take user input for character type
      givenCharacterType = input("\nPlease type V for a villain or H for a hero. \n\t")
      if givenCharacterType == "H" or givenCharacterType == "h":
            characterSubDict["Type"] = "Hero"
            print("Cool, a hero!")
            break
      elif givenCharacterType == "V" or givenCharacterType == "v":
            characterSubDict["Type"] = "Villain"
            print("Oooh, scary, a villain!")
            break
      else:
            print("Sorry, you must type V or H to choose a character type.")
            continue

### builtCharacter = generateCharacter()
### characterSubDict["Name"] = builtCharacter
    # Take user input for name and species; save to keys
    givenCharacterName = input("\nName your character! \n\t")
    characterSubDict["Name"] = givenCharacterName

    givenCharacterSpecies = input("\nWhat species is your character? Ex. Human, Vampire, Ghost, etc. \n\t")
    characterSubDict["Species"] = givenCharacterSpecies

    # Make lists of hero and villain character characteristics
    heroTraits = ["Kind", "Fast", "Brave", "Good leader", "Physically strong", "Innovative", "Creative", "Immortal", "Optimistic", "Great with kids", "Sharpshooter", "Responsible" , "Good at researching", "Historian", "Intelligent", "Selfless", "Loyal", "Super strength", "Can fly", "Good aim", "Charming", "Funny", "Resilient", "Trustworthy", "Wise", "Courageous", "Experienced", "Marytr complex", "Psychic abilities", "Survivor", "Independent", "Team player"]
    villainTraits = ["Cunning", "Sly", "Dishonest", "Manipulative", "Psychic abilities", "Sharpshooter", "Knows magic", "Powers of mind control", "Super strength", "Deceitful", "Cruel", "Knows voodoo", "Ruthless", "Vain", "Vengeful", "Charismatic", "Relentless", "Callous", "Independent", "Spiteful", "Snide", "Vicious", "Bitchy", "Funny", "Bully", "God complex", "Fanatical", "Not from this world", "Ambitious", "Violent", "Nihilistic", "Elaborate planner"]

    print("\nLet's generate some characteristics. ")

    # Create empty list to hold traits
    generatedTraitList = []
    # Trait making loop
    while True:
      # Create random number for generating characteristics
      from random import randint as randomint
      randomCharNum = randomint(0,len(heroTraits)-1)

      # Correctly generate hero or villain trait.
      if characterSubDict["Type"] == "Hero":
        randomTrait = heroTraits[randomCharNum]
      elif characterSubDict["Type"] == "Villain":
        randomTrait = villainTraits[randomCharNum]

      print(f"\nCharacter trait: {randomTrait}")

      # Keep trait loop
      while True:
        keepTrait = input("\nKeep the trait? \n\t(Y to add it to your character \n\t N to get a new trait \n\t X to stop adding traits) \n\t")
        if keepTrait == "Y" or keepTrait == "N" or keepTrait == "X" or keepTrait == "y" or keepTrait == "n" or keepTrait == "x":
          # print("Entered valid letter.")
          break
        else:
          print("\nPlease type Y, N, or X.")
          continue
      if keepTrait == "Y" or keepTrait == "y":
        generatedTraitList.append(randomTrait)
# print(f"Traits = {generatedTraitList}")
        # Another trait loop
        while True:
          anotherTraitYN = input("\nAdd another trait? Y or N. \n\t")
          if anotherTraitYN == "Y" or anotherTraitYN == "N" or anotherTraitYN == "y" or anotherTraitYN == "n":
            # print("You chose a valid Y/N option")
            #  Breaks out of another trait loop
            break
          else:
            print("\nPlease type Y or N.")
            continue
        if anotherTraitYN == "N" or anotherTraitYN == "n":
          # print("You don't want another trait, you want to break the loop.")
          break
        elif anotherTraitYN == "Y" or anotherTraitYN == "N":
          # print("You want another trait, start the loop again!")
          continue

      elif keepTrait == "N":
        # print("Make a new trait.")
        continue

      elif keepTrait == "X":
        # print("You are exiting the trait maker loop?")
        break

    # Get unique entries only by making a set, then make it a list again
    traitSet = set(generatedTraitList)
    uniqueTraitList = list(traitSet)
    if len(uniqueTraitList) == 0:
      characterSubDict["Traits"] = "None"
    else:
      characterSubDict["Traits"] = uniqueTraitList

    # Make lists of possible character weaknesses
    charFlaws = ["Has no physical body", "Lazy", "Afraid of heights", "Gullible", "Obsessive", "Allergic to nuts", "Terrified of water", "Impuslive", "Reckless", "PTSD", "Anxiety", "Anger issues", "Easily manipulated", "Overemotional", "Greedy", "Cursed", "Unlucky", "Masochist", "Low pain tolerance", "Naive", "Pacifist", "Easily distracted", "Scared of needles", "Too trusting", "Overly dramatic", "Stubborn", "Impatient", "Low self esteem", "Terminal illness", "Tragic backstory", "Physically weak", "Scared of fire"]

    print("\nLet's give the character some weaknesses/flaws to keep things interesting. ")

    # Create empty list to hold weaknesses
    generatedFlawsList = []
    # Weakness making loop
    while True:
      # Create random number for generating weaknesses
      from random import randint as randomint
      randomFlawNum = randomint(0,len(charFlaws)-1)

      # Generate a random trait from the list
      randomFlaw = charFlaws[randomFlawNum]

      print(f"\nFlaw: {randomFlaw}")

      # Keep flaw loop
      while True:
        keepFlaw = input("\nKeep this flaw? \n\t(Y to add it to your character \n\t N to get a new flaw \n\t X to stop adding flaws) \n\t")
        if keepFlaw == "Y" or keepFlaw == "N" or keepFlaw == "X" or keepFlaw == "y" or keepFlaw == "n" or keepFlaw == "x":
          break
        else:
          print("\nPlease type Y, N, or X.")
          continue
      if keepFlaw == "Y" or keepFlaw == "y":
        generatedFlawsList.append(randomFlaw)

        # Another flaw loop
        while True:
          anotherFlawYN = input("\nAdd another character flaw? Y or N. \n\t")
          if anotherFlawYN == "Y" or anotherFlawYN == "N" or anotherFlawYN == "y" or anotherFlawYN == "n":
            #  Breaks out of another trait loop
            break
          else:
            print("\nPlease type Y or N.")
            continue
        if anotherFlawYN == "N" or anotherFlawYN == "n":
          break
        elif anotherFlawYN == "Y" or anotherFlawYN == "N":
          continue

      elif keepFlaw == "N":
        continue

      elif keepFlaw == "X":
        break

    # Get unique entries only by making a set, then make it a list again
    flawSet = set(generatedFlawsList)
    uniqueFlawList = list(flawSet)
    if len(uniqueFlawList) == 0:
      characterSubDict["Flaws"] = "None"
    else:
      characterSubDict["Flaws"] = uniqueFlawList

  
      # Return current character, which has 4 keys filled in
# WILL NEED TO ADJUST RANGE WHEN OTHER KEYS ADDED
    print("\nYOUR CHARACTER: ")
    for key in list(characterSubDict)[:5]:
      if(isinstance(characterSubDict[key], list)):
        if len(characterSubDict[key]) == 1:
          print(f"\n{key}: \n\t{characterSubDict[key][0]}")
        else:
          print(f"\n{key}:", end = " ")
          for item in characterSubDict[key]:
            print(f"\n\t- {item}")
      else:
        print(f"\n{key}: \n\t{characterSubDict[key]}")

    # Ask if they would like to keey the character (keepLoop)
    while True:
      happyOrNot = input("\nAre you happy with this character so far? \nType Y for yes or N for no. \n\t")
      if happyOrNot == "Y" or happyOrNot == "y":
         # define marker variable for navigation
        saveToDict = "Y"
        break
# backstory loop was originally here
      elif happyOrNot == "N" or happyOrNot == "n":
        # Try again loop to handle input error
        while True:
          tryAgain = input("\nWould you like to try again? \nType Y for yes or N for no. \n\t")
          if tryAgain == "Y" or tryAgain == "y":
            #redefine navigation variable
            saveToDict = "Y"
            break
          elif tryAgain == "N" or tryAgain == "n":
            # print("We aren't making any more.") this was a marker for me
            #redefine navigation variable
            saveToDict = "N"
            break
          else:
            print("Please type Y or N.")
            continue
        break
      else:
        print("Please type Y or N.")
        continue
    # back in the full character maker loop
    if happyOrNot == "Y" or happyOrNot == "y":
      # Allow them to add a backstory (backstory loop)
      while True:
        inputBackstory = input("\nWould you like to give the character a backstory? \nPlease type Y for yes and N for no. \n\t")
        if inputBackstory == "Y" or inputBackstory == "y":
          # Add 250 word limited backstory loop
          while True:
              givenBackstory = input("\nType a backstory of up to 250 words here: \n\t")
            # .count() found at https://stackoverflow.com/questions/1155617/count-the-number-of-occurrences-of-a-character-in-a-string
              if (givenBackstory.count(" ")+1) < 250:
                characterSubDict["Backstory"] = givenBackstory
                break
              else:
                print("\nPlease limit the backstory to 250 words.")
                continue
          # Print updated character dict with backstory (all keys)
          print("\nWell, that's it for this character. \nHere's what we've got: ")
          for key in list(characterSubDict):
            if(isinstance(characterSubDict[key], list)):
              if len(characterSubDict[key]) == 1:
                print(f"\n{key}: \n\t{characterSubDict[key][0]}")
              else:
                print(f"\n{key}:", end = " ")
                for item in characterSubDict[key]:
                  print(f"\n\t- {item}")
            else:
              print(f"\n{key}: \n\t{characterSubDict[key]}")
          break

        elif inputBackstory == "N" or inputBackstory == "n":
          characterSubDict["Backstory"] = "None"
          print(f"\nNo backstory? Cool, keeping it mysterious!")
          break

        else:
          print("\nSorry, you must type Y or N to proceed.")
          continue
      break
    else:
      if tryAgain == "Y" or tryAgain == "y":
        continue
      if tryAgain == "N" or tryAgain == "n":
        break 

# -------------------------------------------------------------------------

# Create an outer function that allows the creation of multiple characters and saving at the end.
def characterMakerInterface():
# create an empty dictionary to store characters
  personDict = {}
  
  #Character maker loop
  while True:
    # Run generator code to create return single dictionary
    characterBuilder()
    characterMade = characterSubDict

    # Navigation loop; allows for skipping past certain parts that don't apply to user path
    while True:
      if saveToDict == "Y":
      # Ask if they want to save the character (to outer dict)
        userSaveToDict = input("\nWould you like to save this character? \nType Y for yes and N for No. \t")
        # Add to outer dict loop (handle user error)
        while True:
          if userSaveToDict == "Y" or userSaveToDict == "y":
            # Create a key for the character created with a value of the dictionary output
            newKey= f"Character {len(list(personDict.keys()))+1}"
            personDict[newKey] = characterMade
            break
          elif userSaveToDict == "N" or "n":
            print("\nOkay!")
            break
          else:
            print("Please type Y or N.")
            continue
# Breaks out of add to outer dict loop
          break
        break
      elif saveToDict == "N":
        break 

    # Mark with make new or not
    makeNewOrNot = input("\nWould you like to make another character? \nType Y for yes and N for no. \n\t")
    # Make new character loop (handle user error)
    while True:
      if makeNewOrNot == "Y" or makeNewOrNot == "y" or makeNewOrNot == "N" or makeNewOrNot == "n":
        break
      else:
        print("Please type Y or N.")
        continue
    
# THIS IS BACK IN FULL CHAR MAKER LOOP
    if makeNewOrNot == "Y" or makeNewOrNot == "y":
      continue
    if makeNewOrNot == "N" or makeNewOrNot == "n":
      break

  # See if outer dict is empty; if not, do not write file.
  if len(list(personDict.keys())) == 0:
    print("\nOkay, maybe you'll come back to make characters some other time. \nThanks for trying my character creator!")
  else:
  # Write json file loop user choice & error handling)
    while True:
      # Ask if they would like to download character file
      saveFileYN = input("\nWould you like to save a file of your character(s) information? \nType Y for yes or N for No. \n\t")
      if saveFileYN == "Y" or saveFileYN == "y":
        # Accept user input for naming the file
        fileNameInput = input("\nPlease name your file. \t")
        fileName = f"{fileNameInput}.json"
        filePath = fileName
        # Import json to use for file writing
        import json
        # Write the dictionary to a file
        with open(filePath, "w") as json_write:
          json.dump(personDict, json_write, indent = 2)
          print(f"\nFile written! \n\nYou can download your file, {filePath}, by clicking the folder icon in the left sidebar and clicking the three dots next to your file. \n(If it's not there, click the folder icon with the refresh arrow at the top of the Files window.)")
          print("Thank you for using my character creator! :^)")
          return
      elif saveFileYN == "N" or saveFileYN == "n":
        print("\nOkay, thank you for using my character creator!")
        return
      else:
        print("\nSorry, you must type Y or N to proceed.")
        continue
    
# characterBuilder()
characterMakerInterface()
```
